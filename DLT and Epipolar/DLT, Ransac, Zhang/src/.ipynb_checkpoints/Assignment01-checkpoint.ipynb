```python
%matplotlib inline

def show2DProjections(pts, path):
    fig = plt.figure(figsize=(10, 20))
    img = cv2.cvtColor(cv2.imread(path), cv2.COLOR_BGR2RGB)
    plt.imshow(img)

    n = len(pts)
    colors = np.random.rand(n)
    area = (5 * np.ones(n))**2 

    plt.scatter(pts[:, 0], pts[:, 1], c=colors, s=area)
    plt.show()
```

```python
def DLT(X, x):
    A = []
    
    for (xi, yi), (Xi, Yi, Zi) in zip(x, X):
        A.append([Xi, Yi, Zi, 1, 0, 0, 0, 0, -Xi * xi, -Yi * xi, -Zi * xi, -xi])
        A.append([0, 0, 0, 0, Xi, Yi, Zi, 1, -Xi * yi, -Yi * yi, -Zi * yi, -yi])
    
    A = np.array(A)
    U, D, VT = np.linalg.svd(A)
    P = VT[-1, :].reshape(3, 4)
    P /= P[-1, -1]
    return P
```

```python
def getPts(P, X):
    homo3d = np.vstack([X.T, np.ones(X.shape[0])])
    reproj = P @ homo3d
    reproj = reproj / reproj[2, :]
    reproj = np.array(reproj[:2, :])
    return reproj.T
```

```python
def getErr(P, X, x):
    pts = getPts(P, X)
    diff = x - pts
    err = np.linalg.norm(diff, axis=1)
    mse = np.mean(err ** 2)
    rmse = np.sqrt(mse)
    return [mse, rmse]
```

```python
def calibTest(P):
    test3D = np.array([[108, 72, 0],
                       [144, 72, 0],
                       [72, 0, 72],
                       [108, 0, 72],
                       [108, 0, 108]])
    test2D = getPts(P, test3D)
    show2DProjections(test2D, path)
```

### 2. RANSAC

```python
def RANSAC(X, x, itr, thresh = 0.1):
    maxIn = -1
    opP = np.ones((3, 4))
    
    for i in range(itr):
        inds = np.random.randint(x.shape[0], size = 10)
        P = DLT(X[inds], x[inds])
        reproj = getPts(P, X)
        reprojErr = np.sqrt(np.mean(np.square(x - reproj), axis = 0))
        inliers = np.where(reprojErr < thresh)[0]
        
        if maxIn < len(inliers):
            maxIn = len(inliers)
            opP = P
            
    print('Number of Inliers = ', maxIn)
    return opP
```

```python
opP = RANSAC(x3d, x2d, 500, 10)
print("Optimized P = \n", opP)
calibTest(opP)
```

### 3. Decomposition

```python
def getKRT(P):
    M = P[:, :-1]
    invM = np.linalg.inv(M)
    h = P[:, -1].reshape(3, 1)
    
    RT, invK = np.linalg.qr(invM)
    K = np.linalg.inv(invK)
    R = RT.T
    K = K / K[-1, -1]
    t = -invM @ h
    return K, R, t
```

```python
K, R, t = getKRT(P)
print("K = \n", K, "\nR = \n", R, "\nt = \n", t)
```

```python
opK, opR, opt = getKRT(opP)
print("For optimized P: \nK = \n", opK, "\nR = \n", opR, "\nt = \n", opt)
```

### 4. Radial Distortion Correction

```python
img = cv2.cvtColor(cv2.imread(path), cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

tempK = K
tempK[0][1] = 0
if tempK[0][0] < 0:
    tempK[0][0] = -tempK[0][0]
if tempK[1][1] < 0:
    tempK[1][1] = -tempK[1][1]

ret, newK, dist, newR, newT = cv2.calibrateCamera([x3d.astype('float32')], [x2d.astype('float32')], (gray.shape[1], gray.shape[0]), tempK, None, None, flags=(cv2.CALIB_USE_INTRINSIC_GUESS))
h, w = gray.shape[:2]
newEstK, roi = cv2.getOptimalNewCameraMatrix(tempK, dist, (w, h), 1, (w, h))
M = K @ R
C = K @ R @ t
newP = np.hstack([M, C])
fixedImg = cv2.undistort(img, newK, dist, None, newEstK)

plt.figure()
plt.imshow(img)
plt.title("Distorted Image")

plt.figure()
plt.imshow(fixedImg)
plt.title("Fixed/Undistorted Image")
plt.savefig("../images/Undistorted Image.jpg")
plt.show()
print("Distortion Coefficients = \n", dist)
```
